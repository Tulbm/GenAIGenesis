```python
# import the necessary libraries
from gensim import corpora, models
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import nltk
import spacy
```

```python
text = """
Meet Alex Reed, a vibrant soul with an insatiable curiosity for the world's mysteries and a heart filled with wanderlust. Born and raised in the bustling city of Toronto, Alex has always found solace in the pages of books, leading them to develop a profound love for literature and history. Weekends for Alex are often spent exploring the nooks and crannies of antique bookshops or getting lost in the vast collections of the city's museums. 

Alex's passion for history isn't just confined to books; it extends to a deep-seated love for travel, especially to places with rich historical backgrounds. Rome, with its ancient ruins and timeless stories, is at the top of their travel bucket list. However, Alex isn't just about the past; they're equally fascinated by the future, particularly the possibilities within the realm of technology and AI, spending evenings coding or dabbling in digital art projects.

Despite their diverse interests, Alex cherishes simplicity in daily life. They find joy in the small things: a perfectly brewed cup of coffee in the morning, the tranquil silence of early dawn, and the rhythmic sound of rain against the window. Alex is also an advocate for mental health awareness, often volunteering at local community centers to help those in need of support and understanding.

In a world that's constantly moving, Alex Reed stands as a testament to the beauty of merging past interests with future dreams, all while staying grounded in the present moment's simple pleasures.
"""
text = text.translate(str.maketrans('', '', string.punctuation))
```

```python
text = text.replace("\n", " ")
```

```python
text = text.lower()
```

```python
text
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp(text)
nouns = [token.text for token in doc if token.pos_ in ["NOUN", "PROPN"]]
```

```python
nouns = set(nouns)
```

```python
nouns
```

```python
# create a dictionary from the text data
dictionary = corpora.Dictionary([nouns])
```

```python
dictionary
```

```python
# create a corpus from the text data
corpus = [dictionary.doc2bow(nouns)]
```

```python
corpus
```

```python
# train the LDA model on the corpus
ldamodel = models.LdaModel(corpus, num_topics=3, id2word=dictionary)

# extract the topics from the model
topics = ldamodel.print_topics(num_topics=3, num_words=1)

# print the topics
topic_list = []
for topic in topics:
    topic_list.append(topic[1].split('"')[1])
```
